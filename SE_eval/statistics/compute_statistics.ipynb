```python
# Display results in rows
import pandas as pd
from IPython.display import display, HTML
import numpy as np

def display_df(params):
    metric_columns = ["SI-SDR", "PESQ", "STOI", "MOS_SIG", "MOS_BAK", "MOS_OVR"]
    combined_results = pd.DataFrame(columns=["SI-SDR ± SE", "PESQ ± SE", "STOI ± SE"])

    for key, filepath in params.items():
        df = pd.read_csv(filepath)

        # Filter the DataFrame
        condition = (df['Noise SNR'] > -10) & (df['Noise SNR'] < 10) # & (df['Noise Type'] != 'Street') & (df['Noise Type'] != 'Cafe')
        df_filtered = df[condition]
        # print(len(df_filtered))
        # Calculate average and standard error
        average = df_filtered[metric_columns].mean()
        standard_error = df_filtered[metric_columns].std() / np.sqrt(len(df_filtered))

        # Round the values to two decimal places
        average = average.round(2)
        standard_error = standard_error.round(2)
        
        # Create a new DataFrame for the data frame's metrics and their standard errors
        df_row = pd.DataFrame({f"{metric} ± SE": [f"{average[metric]} ± {standard_error[metric]}"] for metric in metric_columns})
        df_row.index = [key]

        # Append the data frame's row to the combined results DataFrame
        combined_results = pd.concat([combined_results, df_row])

    # Display the combined results
    display(combined_results)
```

```python
# Display results side-by-side
import pandas as pd
from IPython.display import display, HTML
import numpy as np

def display_df(params):
    N = len(params)
    table_html_list = []  # Initialize an empty list to store the HTML representation of each DataFrame

    for ind_i, key_i in enumerate(params.keys()):
        # Read the CSV file into a DataFrame
        df_i = pd.read_csv(params[key_i])

        # Define the filtering condition using boolean indexing
        condition = (df_i['Noise SNR'] > -10) & (df_i['Noise SNR'] < 10) & (df_i['Noise Type'] != 'Street') & (df_i['Noise Type'] != 'Cafe')

        # Apply the condition to the DataFrame
        df_i = df_i[condition]

        # Step 1: Calculate the average of each column in the DataFrame
        average_row = df_i.mean(numeric_only=True)
        average_row['Noise SNR'] = 'Average'  # Adding the label "Average" to the 'Noise SNR' column

        sde_row = df_i.std(numeric_only=True)/np.sqrt(df_i.shape[0]) # Standard error
        sde_row['Noise SNR'] = 'Standard error'  # Adding the label "Average" to the 'Noise SNR' column

        # Step 2: Create a new DataFrame with the average row
        df_avg = pd.DataFrame([average_row])
        df_sde = pd.DataFrame([sde_row])

        # Step 3: Concatenate the average row DataFrame to the original DataFrame
        df_i = pd.concat([df_i, df_avg, df_sde], ignore_index=True)

        df_i_styler = df_i.groupby(['Noise SNR']).mean(numeric_only=True).style.set_table_attributes("style='display:inline'").set_caption(key_i)
        table_html_list.append(df_i_styler.to_html())  # Append the HTML representation of the DataFrame to the list

    # Combine the HTML representations with some margin between them
    combined_html = " ".join(f"<div style='margin-right: 20px; display: inline-block;'>{table_html}</div>" for table_html in table_html_list)

    # Display the combined HTML
    display(HTML(combined_html))
```
